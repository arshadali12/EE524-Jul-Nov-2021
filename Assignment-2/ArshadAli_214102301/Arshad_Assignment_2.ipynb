### Continuous Distributions

#### Uniform Distribution

```python
def uniform_pdf(a,b,x):
    if(x<=b and x>=a):
        return 1/(b-a)
    else:
        return 0;
    
def uniform_cdf(a,b,x):
    if x<a:
        return 0
    elif x>b:
        return 1
    else:
        return (x-a)/(b-a)
    
def uniform_mean_median(a,b):
    # mean and median same for uniform
    return 0.5*(a+b)
def uniform_var(a,b):
    return ((b-a)**2)/12

x_val = np.linspace(-5,5,1000)
y_val1 = np.array([uniform_pdf(-4,-1,x) for x in x_val])
y_val2 = np.array([uniform_pdf(2,3,x) for x in x_val])
y_val3 = np.array([uniform_pdf(1,4,x) for x in x_val])

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15,10))


ax[0][0].plot(x_val,y_val1,label='PDF a=-4  b=-1')
ax[0][0].axvline(x=uniform_mean_median(-4,-1),color='red',linestyle='--',alpha=0.5,label='Mean and Median')
ax[0][0].axvline(x=uniform_var(-4,-1),color='green',linestyle='-.',label='Variance')
ax[0][0].legend()
ax[0][0].grid()
ax[1][0].plot(x_val, np.array([uniform_cdf(-4,-1,x) for x in x_val])  , label="CDF")
ax[1][0].legend()
ax[1][0].grid()

ax[0][1].plot(x_val,y_val2,label='PDF a=2  b=3')
ax[0][1].axvline(x=uniform_mean_median(2,3),color='red',linestyle='--',alpha=0.5,label='Mean and Median')
ax[0][1].axvline(x=uniform_var(2,3),color='green',linestyle='-.',label='Variance')
ax[0][1].legend()
ax[0][1].grid()
ax[1][1].plot(x_val, np.array([uniform_cdf(2,3,x) for x in x_val])  , label="CDF")
ax[1][1].legend()
ax[1][1].grid()

ax[0][2].plot(x_val,y_val3,label='PDF a=1  b=4')
ax[0][2].axvline(x=uniform_mean_median(1,4),color='red',linestyle='--',alpha=0.5,label='Mean and Median')
ax[0][2].axvline(x=uniform_var(1,4),color='green',linestyle='-.',label='Variance')
ax[0][2].legend()
ax[0][2].grid()
ax[1][2].plot(x_val, np.array([uniform_cdf(1,4,x) for x in x_val])  , label="CDF")
ax[1][2].legend()
ax[1][2].grid()


plt.suptitle("Uniform Distribution")
```

#### Exponential Distribution

```python
def exponential_pdf(lam , x ):
    # Exponential PDF = lambda* exp(-lambda*x)
    return lam*np.exp(-lam*x)

def exponential_cdf(lam,x):
    return 1-exp(-lam*x)

def exponential_mean(lam):
    return 1/lam
def exponential_median(lam):
    return 0.693/lam
def exponential_var(lam):
    return 1/(lam**2)

x_val = np.linspace(-5,5,1000)
y_val1 = np.array([exponential_pdf(0.5,x) for x in x_val])
y_val2 = np.array([exponential_pdf(1,x) for x in x_val])
y_val3 = np.array([exponential_pdf(1.5,x) for x in x_val])

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15,10))


ax[0][0].plot(x_val,y_val1,label='PDF lambda = 0.5')
ax[0][0].axvline(x=exponential_mean(0.5),color='red',linestyle='--',alpha=0.5,label='Mean')
ax[0][0].axvline(x=exponential_median(0.5),color='black',linestyle='-',alpha=0.5,label='Median')
ax[0][0].axvline(x=exponential_var(0.5),color='green',linestyle='-.',label='Variance')
ax[0][0].legend()
ax[0][0].grid()
ax[1][0].plot(x_val,np.array([exponential_cdf(0.5,x) for x in x_val]),label='CDF lambda = 0.5')
ax[1][0].legend()
ax[1][0].grid()


ax[0][1].plot(x_val,y_val2,label='lambda = 1.0')
ax[0][1].axvline(x=exponential_mean(1),color='red',linestyle='--',alpha=0.5,label='Mean')
ax[0][1].axvline(x=exponential_median(1),color='black',linestyle='-',alpha=0.5,label='Median')
ax[0][1].axvline(x=exponential_var(1),color='green',linestyle='-.',label='Variance')
ax[0][1].set_ylim(0,5)
ax[0][1].legend()
ax[0][1].grid()
ax[1][1].plot(x_val,np.array([exponential_cdf(1,x) for x in x_val]),label='CDF lambda = 1.0')
ax[1][1].legend()
ax[1][1].grid()


ax[0][2].plot(x_val,y_val3,label='lambda = 1.5')
ax[0][2].axvline(x=exponential_mean(1.5),color='red',linestyle='--',alpha=0.5,label='Mean')
ax[0][2].axvline(x=exponential_median(1.5),color='black',linestyle='-',alpha=0.5,label='Median')
ax[0][2].axvline(x=exponential_var(1.5),color='green',linestyle='-.',label='Variance')
ax[0][2].set_ylim(0,5)
ax[0][2].legend()
ax[0][2].grid()
ax[1][2].plot(x_val,np.array([exponential_cdf(1.5,x) for x in x_val]),label='CDF lambda = 1.5')
ax[1][2].legend()
ax[1][2].grid()


plt.suptitle("Exponential Distribution")
```

#### Gaussian Distribution

```python
def gaussian_pdf(mu,sigma , x ):
    return exp(-0.5*(((x-mu)/sigma)**2))/(sigma*((2*3.14)**0.5))

x_val = np.linspace(-5,5,1000)
y_val1 = np.array([gaussian_pdf(0,1,x) for x in x_val])
y_val2 = np.array([gaussian_pdf(-2,2,x) for x in x_val])
y_val3 = np.array([gaussian_pdf(2,1,x) for x in x_val])


fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15,10))


ax[0][0].axvline(x=0,color='red',linestyle='--',alpha=0.5,label='Mean,Median,Mode')
ax[0][0].axvline(x=1,color='green',linestyle='-.',label='Variance') # Variance = Sigma^2
ax[0][0].plot(x_val,y_val1,label='mu=0 , sigma=1')
ax[0][0].legend()

ax[0][1].axvline(x=-2,color='red',linestyle='--',alpha=0.5,label='Mean,Median,Mode')
ax[0][1].axvline(x=2**2,color='green',linestyle='-.',label='Variance')
ax[0][1].plot(x_val,y_val2,label='mu=-2 , sigma=2')
ax[0][1].legend()

ax[0][2].axvline(x=2,color='red',linestyle='--',alpha=0.5,label='Mean,Median,Mode')
ax[0][2].axvline(x=1,color='green',linestyle='-.',label='Variance')
ax[0][2].plot(x_val,y_val3,label='mu=2 , sigma=1')
ax[0][2].legend()
```

### Discrete Distribution

#### (a) The received bit data follows bernoulli distribution
#### PMF of Bernoulli Distribution

```python
def bernoulli_pmf(p , x ):
    if x==0:
        return 1-p
    elif x==1:
        return p
    else:
        return 0
    
def bernoulli_mean(p):
    # Mean of bernoulli distribution is p
    return p

def bernoulli_var(p):
    # Variance of bernoulli distribution is 1-p
    return 1-p

x_val=np.linspace(-1,2,1000)
y_val1= np.array([bernoulli_pmf(0.1,x) for x in x_val])
y_val2= np.array([bernoulli_pmf(0.5,x) for x in x_val])
y_val3= np.array([bernoulli_pmf(0.8,x) for x in x_val])

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15,5))


ax[0].plot(x_val,y_val1,label='p = 0.1')
ax[0].axvline(x=bernoulli_mean(0.1),color='red',linestyle='--',alpha=0.5,label='Mean')
ax[0].axvline(x=bernoulli_var(0.1),color='green',linestyle='-.',label='Variance')

ax[1].plot(x_val,y_val2,label='p = 0.5')
ax[1].axvline(x=bernoulli_mean(0.5),color='red',linestyle='--',alpha=0.5,label='Mean')
ax[1].axvline(x=bernoulli_var(0.5),color='green',linestyle='-.',label='Variance')

ax[2].plot(x_val,y_val3,label='p = 0.8')
ax[2].axvline(x=bernoulli_mean(0.8),color='red',linestyle='--',alpha=0.5,label='Mean')
ax[2].axvline(x=bernoulli_var(0.8),color='green',linestyle='-.',label='Variance')


for axes in ax:
    axes.legend()
    axes.grid()
    axes.set_ylim(0,1)

plt.suptitle("PMF of Bernoulli Distribution")
```

#### (b) This time recieved data follows Binomial distribution, and we are doing trial for n=20 times and probability of success p=0.4
#### PMF of Binomial Distribution 

```python
from math import comb

x_val=np.arange(0,21,0.01)

def binomial_pmf(p ,n, x ):
        if(x.is_integer()):
            return comb(n,int(x))*(p**int(x))*((1-p)**int(x))
        else:
            return 0

def binomial_mean(p,n):
    # Mean of binomial distribution is np
    return n*p

def binomial_var(p,n):
    # Variance of bernoulli distribution is np(1-p)
    return n*p*(1-p)

y_val= np.array([binomial_pmf(0.4,20,x) for x in x_val])

fig, ax = plt.subplots()


ax.plot(x_val,y_val,label='p = 0.4')
ax.grid()
ax.axvline(x=binomial_mean(0.4,20),color='red',linestyle='--',alpha=0.5,label='Mean')
ax.axvline(x=binomial_var(0.4,20),color='green',linestyle='-.',label='Variance')
ax.legend()
```

#### (c) Received data follows geometric distribution.
#### PMF of Geometric Distribution 

```python
x_val=np.arange(0,11,0.01)
def geometric_pmf(p , x ):
    if(x.is_integer()):
        return p*((1-p)**(x-1))
    
    else:
        return 0
    
def geometric_mean(p):
    # Mean of bernoulli distribution is 1/p
    return 1/p

def geometric_var(p):
    # Variance of bernoulli distribution is (1-p)/p^2
    return (1-p)/(p**2)

y_val1= np.array([geometric_pmf(0.4,x) for x in x_val])
y_val2= np.array([geometric_pmf(0.5,x) for x in x_val])
y_val3= np.array([geometric_pmf(0.8,x) for x in x_val])

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15,5))


ax[0].plot(x_val,y_val1,label='p = 0.4')
ax[0].axvline(x=geometric_mean(0.4),color='red',linestyle='--',alpha=0.5,label='Mean')
ax[0].axvline(x=geometric_var(0.4),color='green',linestyle='-.',label='Variance')
ax[0].grid()
ax[0].legend()


ax[1].plot(x_val,y_val2,label='p = 0.5')
ax[1].axvline(x=geometric_mean(0.5),color='red',linestyle='--',alpha=0.5,label='Mean')
ax[1].axvline(x=geometric_var(0.5),color='green',linestyle='-.',label='Variance')
ax[1].grid()
ax[1].legend()


ax[2].plot(x_val,y_val3,label='p = 0.8')
ax[2].axvline(x=geometric_mean(0.8),color='red',linestyle='--',alpha=0.5,label='Mean')
ax[2].axvline(x=geometric_var(0.8),color='green',linestyle='-.',label='Variance')
ax[2].grid()
ax[2].legend()




plt.suptitle("PMF of Geometric Distribution")
```

#### (d) Random Variable X follows Poisson Distribution
#### PMF of Poisson Distribution 

```python
from math import factorial
from math import exp
alpha = 5
x_val=np.arange(0,11,0.01)
def poisson_pmf(alpha , x ):
    if(x.is_integer()):
        return ((alpha**x)*(exp(-alpha)))/factorial(x)
    
    else:
        return 0
    
def poisson_mean_var(alpha):
    return alpha

y_val = np.array([poisson_pmf(5,x) for x in x_val])
fig, ax = plt.subplots( figsize=(15,5))


ax.plot(x_val,y_val,label='alpha = 5')
ax.axvline(x=poisson_mean_var(5),color='red',linestyle='--',alpha=0.5,label='Mean')
ax.axvline(x=poisson_mean_var(5),color='green',linestyle='-.',label='Variance')

ax.grid()
ax.legend()
ax.set_title("Poisson Distribution")

```
